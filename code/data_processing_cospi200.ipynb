### PER Nan값 채우기

```python
# 크롤링한 종가 데이터 csv 불러오기
# 절대경로 사용
df_price = pd.read_csv('/Users/sungsupark/Desktop/2025-1 수업자료/소프트웨어융합캡스톤디자인/csv/kospi200_year_end.csv')
```

```python
# 1. 'PER(배)'가 NaN인 행만 선택한다.
df_nan_per = df[df['PER(배)'].isna()].copy()

# 2. 기업명과 연도를 분리한다. (예시: 'SK하이닉스(2023/12)' → 기업명: SK하이닉스, 연도: 2023)
df_nan_per[['기업명_추출', '연도']] = df_nan_per['기업명'].str.extract(r'(.+)\((\d{4})')

# 3. 연도에 맞는 종가를 찾는다. df_stock에서 해당 연도, 해당 기업명의 값을 가져와서 나눔
def get_stock_price(row):
    year = int(row['연도'])
    company = row['기업명_추출']
    try:
        # 연도별로 행이 2020년은 index 0부터 시작 → 2020: index 0, 2021: index 1, ...
        stock_price = df_price.loc[year - 2020, company]
        return stock_price
    except:
        return np.nan

# 주가 구하기
df_nan_per['주가'] = df_nan_per.apply(get_stock_price, axis=1)

# PER 계산: 주가 / EPS
df_nan_per['계산_PER'] = df_nan_per['주가'] / df_nan_per['EPS(원)']

# 원래 df_financial에 채워넣는다.
for idx, row in df_nan_per.iterrows():
    if not np.isnan(row['계산_PER']):
        df.at[idx, 'PER(배)'] = row['계산_PER']

# 예시로 PER이 Nan값이었던 SK하이닉스를 확인해본다.
df[df['기업명'].str.contains('SK하이닉스')]  
```

### PBR값 채우기 (KG모빌리티)

```python
# 'KG모빌리티'의 2020년 연말 종가는 13140원이다.
# 'KG모빌리티'의 2021년 연말 종가는 13140원이다.

# PBR = 주가 / 주당순자산(BPS)

# 'KG모빌리티(2020/12)'의 주당순자산(BPS)는 -5706.0원이다.
# 'KG모빌리티(2021/12)'의 주당순자산(BPS)는 -6414.0원이다.
pbr_2020 = 13140.0 / -5706.0
pbr_2021 = 13140.0 / -6414.0

# '기업명'이 'KG모빌리티(2020/12)'인 행의 'PBR(배)' 값을 계산하여 삽입한다.
# '기업명'이 'KG모빌리티(2021/12)'인 행의 'PBR(배)' 값을 계산하여 삽입한다.
df.loc[df['기업명'] == 'KG모빌리티(2020/12)', 'PBR(배)'] = pbr_2020
df.loc[df['기업명'] == 'KG모빌리티(2021/12)', 'PBR(배)'] = pbr_2021
```

### 'ROE', 'ROA' Nan 값 채우기

```python
# 1. 'ROE(%)'가 NaN인 행만 선택한다.
df_nan_roe = df[df['ROE(%)'].isna()].copy()

# 2. 기업명과 연도 분리한다.
df_nan_roe[['기업명_추출', '연도']] = df_nan_roe['기업명'].str.extract(r'(.+)\((\d{4})')

# 3. ROE 계산: (당기순이익 / 자본총계) * 100
df_nan_roe['계산_ROE'] = (df_nan_roe['당기순이익'] / df_nan_roe['자본총계']) * 100

# 원래 df에 채워넣는다.
for idx, row in df_nan_roe.iterrows():
    if not np.isnan(row['계산_ROE']):
        df.at[idx, 'ROE(%)'] = row['계산_ROE']

# 4. 'ROA(%)'가 NaN인 행만 선택한다.
df_nan_roa = df[df['ROA(%)'].isna()].copy()

# 5. 기업명과 연도 분리한다.
df_nan_roa[['기업명_추출', '연도']] = df_nan_roa['기업명'].str.extract(r'(.+)\((\d{4})')

# 6. ROA 계산: (당기순이익 / 자산총계) * 100
df_nan_roa['계산_ROA'] = (df_nan_roa['당기순이익'] / df_nan_roa['자산총계']) * 100

# 원래 df에 채워넣는다.
for idx, row in df_nan_roa.iterrows():
    if not np.isnan(row['계산_ROA']):
        df.at[idx, 'ROA(%)'] = row['계산_ROA']
```

### (기존 방법)5년치 데이터 통합 (가중치 사용)

### 가중치를 어떻게 설정해야하지?

```python
# '기업명' 열에서 연도를 분리하고, 가중치를 계산한 후 가중평균을 적용하여 데이터를 통합합니다.

# 1. '기업명'에서 연도를 추출합니다.
df['연도'] = df['기업명'].str.extract(r'\((\d{4})')[0].astype(int)

# 2. 가중치를 계산합니다.
# 가장 최근 연도에 높은 가중치를 부여합니다.
max_year = df['연도'].max()
df['가중치'] = df['연도'].apply(lambda x: x - (max_year - 5) + 1)

# 3. '기업명'에서 연도를 제거하여 그룹화할 기준을 만듭니다.
df['기업명_그룹'] = df['기업명'].str.extract(r'(.+)\(')[0]

# 4. 가중평균을 계산하는 함수입니다.
def weighted_average(group):
    weights = group['가중치']
    return group.drop(columns=['기업명', '연도', '가중치', '기업명_그룹']).multiply(weights, axis=0).sum() / weights.sum()

# 5. 그룹별로 가중평균을 계산합니다.
final_df = df.groupby('기업명_그룹').apply(weighted_average).reset_index()

# 6. 결과를 확인합니다.
final_df
```

### Z-score가 3이상으로 나온 삼성전자와 SK하이닉스 배제

```python
# '기업명_그룹' 열에서 '삼성전자'와 'SK하이닉스'를 가진 행 제거
filtered_df = final_df[~final_df['기업명_그룹'].isin(['삼성전자', 'SK하이닉스'])]

# 제거 후 크기 확인
print(f"제거 후 데이터프레임 크기: {filtered_df.shape}")
print(f"제거된 행 수: {final_df.shape[0] - filtered_df.shape[0]}")

# 원본 데이터프레임을 변경하고 싶다면:
final_df = final_df[~final_df['기업명_그룹'].isin(['삼성전자', 'SK하이닉스'])]
```

### (새로운 방법) 보완 분석 (Window사용)

### Window(3년)을 나누어 분석을 진행

```python
# 1. '기업명'에서 연도 추출 - 아직 '연도' 열이 없음
if '연도' not in df.columns:
    df['연도'] = df['기업명'].str.extract(r'\((\d{4})')[0].astype(int)

# 2. '기업명'에서 연도를 제거하여 그룹화할 기준을 만듦(아직 '기업명_그룹' 열이 없음)
if '기업명_그룹' not in df.columns:
    df['기업명_그룹'] = df['기업명'].str.extract(r'(.+)\(')[0]

# Window1: 2020-2022년 데이터
# 1. 2020-2022 기간의 데이터만 필터링
df_window1 = df[df['연도'].between(2020, 2022)].copy()

# 2. 가중치 재계산 (2022년에 가장 큰 가중치)
df_window1['가중치'] = df_window1['연도'].apply(lambda x: x - 2019)  # 2020년=1, 2021년=2, 2022년=3

# 3. 가중평균 계산
def weighted_average(group):
    weights = group['가중치']
    return group.drop(columns=['기업명', '연도', '가중치', '기업명_그룹']).multiply(weights, axis=0).sum() / weights.sum()

# 4. 기업별로 가중평균 계산
df_2020_2022 = df_window1.groupby('기업명_그룹').apply(weighted_average).reset_index()
print(f"Window 2020-2022 데이터 크기: {df_2020_2022.shape}")

# Window2: 2021-2023년 데이터
# 1. 2021-2023 기간의 데이터만 필터링
df_window2 = df[df['연도'].between(2021, 2023)].copy()

# 2. 가중치 재계산 (2023년에 가장 큰 가중치)
df_window2['가중치'] = df_window2['연도'].apply(lambda x: x - 2020)  # 2021년=1, 2022년=2, 2023년=3

# 3. 기업별로 가중평균 계산
df_2021_2023 = df_window2.groupby('기업명_그룹').apply(weighted_average).reset_index()
print(f"Window 2021-2023 데이터 크기: {df_2021_2023.shape}")

# Window3: 2022-2024년 데이터
# 1. 2022-2024 기간의 데이터만 필터링
df_window3 = df[df['연도'].between(2022, 2024)].copy()

# 2. 가중치 재계산 (2024년에 가장 큰 가중치)
df_window3['가중치'] = df_window3['연도'].apply(lambda x: x - 2021)  # 2022년=1, 2023년=2, 2024년=3

# 3. 기업별로 가중평균 계산
df_2022_2024 = df_window3.groupby('기업명_그룹').apply(weighted_average).reset_index()
print(f"Window 2022-2024 데이터 크기: {df_2022_2024.shape}")

# 데이터 확인 (상위 5개 행만)
print("\n----- Window 2020-2022 상위 5개 행 -----")
print(df_2020_2022.head())

print("\n----- Window 2021-2023 상위 5개 행 -----")
print(df_2021_2023.head())

print("\n----- Window 2022-2024 상위 5개 행 -----")
print(df_2022_2024.head())
```

### Z-score 3이상인 삼성전자, SK하이닉스 배제 및 최종 데이터 저장

```python
# 각 Window 데이터에서 삼성전자, SK하이닉스 제거
# Window 2020-2022
df_2020_2022 = df_2020_2022[~df_2020_2022['기업명_그룹'].isin(['삼성전자', 'SK하이닉스'])]

# Window 2021-2023
df_2021_2023 = df_2021_2023[~df_2021_2023['기업명_그룹'].isin(['삼성전자', 'SK하이닉스'])]

# Window 2022-2024
df_2022_2024 = df_2022_2024[~df_2022_2024['기업명_그룹'].isin(['삼성전자', 'SK하이닉스'])]

# 최종 크기 확인
print(f"Window 2020-2022 최종 크기: {df_2020_2022.shape}")
print(f"Window 2021-2023 최종 크기: {df_2021_2023.shape}")
print(f"Window 2022-2024 최종 크기: {df_2022_2024.shape}")

# 결과를 CSV 파일로 저장
df_2020_2022.to_csv('financial_data_window_2020_2022.csv', index=False, encoding='utf-8-sig')
df_2021_2023.to_csv('financial_data_window_2021_2023.csv', index=False, encoding='utf-8-sig')
df_2022_2024.to_csv('financial_data_window_2022_2024.csv', index=False, encoding='utf-8-sig')

print("각 Window 데이터가 CSV 파일로 저장되었습니다.")
```
